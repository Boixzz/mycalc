```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QHBoxLayout, QLineEdit, QPushButton, QLabel

class Calculator(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Calculator")
        self.initUI()

    def initUI(self):
        # Create the display widget
        self.display = QLineEdit()
        self.display.setAlignment(Qt.AlignRight)
        self.display.setReadOnly(True)
        self.display.setStyleSheet("font-size: 20px;")

        # Create the buttons
        self.btn1 = QPushButton("1")
        self.btn2 = QPushButton("2")
        self.btn3 = QPushButton("3")
        self.btn4 = QPushButton("4")
        self.btn5 = QPushButton("5")
        self.btn6 = QPushButton("6")
        self.btn7 = QPushButton("7")
        self.btn8 = QPushButton("8")
        self.btn9 = QPushButton("9")
        self.btn0 = QPushButton("0")
        self.btnplus = QPushButton("+")
        self.btnminus = QPushButton("-")
        self.btnmultiply = QPushButton("*")
        self.btndivide = QPushButton("/")
        self.btnequal = QPushButton("=")
        self.btnclear = QPushButton("C")

        # Create the layout
        vbox = QVBoxLayout()
        hbox1 = QHBoxLayout()
        hbox2 = QHBoxLayout()
        hbox3 = QHBoxLayout()
        hbox4 = QHBoxLayout()

        # Add the widgets to the layout
        hbox1.addWidget(self.display)
        vbox.addLayout(hbox1)

        hbox2.addWidget(self.btn1)
        hbox2.addWidget(self.btn2)
        hbox2.addWidget(self.btn3)
        hbox2.addWidget(self.btnplus)
        vbox.addLayout(hbox2)

        hbox3.addWidget(self.btn4)
        hbox3.addWidget(self.btn5)
        hbox3.addWidget(self.btn6)
        hbox3.addWidget(self.btnminus)
        vbox.addLayout(hbox3)

        hbox4.addWidget(self.btn7)
        hbox4.addWidget(self.btn8)
        hbox4.addWidget(self.btn9)
        hbox4.addWidget(self.btnmultiply)
        vbox.addLayout(hbox4)

        hbox5 = QHBoxLayout()
        hbox5.addWidget(self.btnclear)
        hbox5.addWidget(self.btn0)
        hbox5.addWidget(self.btnequal)
        hbox5.addWidget(self.btndivide)
        vbox.addLayout(hbox5)

        # Connect the buttons to the slots
        self.btn1.clicked.connect(lambda: self.onButtonClicked("1"))
        self.btn2.clicked.connect(lambda: self.onButtonClicked("2"))
        self.btn3.clicked.connect(lambda: self.onButtonClicked("3"))
        self.btn4.clicked.connect(lambda: self.onButtonClicked("4"))
        self.btn5.clicked.connect(lambda: self.onButtonClicked("5"))
        self.btn6.clicked.connect(lambda: self.onButtonClicked("6"))
        self.btn7.clicked.connect(lambda: self.onButtonClicked("7"))
        self.btn8.clicked.connect(lambda: self.onButtonClicked("8"))
        self.btn9.clicked.connect(lambda: self.onButtonClicked("9"))
        self.btn0.clicked.connect(lambda: self.onButtonClicked("0"))
        self.btnplus.clicked.connect(lambda: self.onButtonClicked("+"))
        self.btnminus.clicked.connect(lambda: self.onButtonClicked("-"))
        self.btnmultiply.clicked.connect(lambda: self.onButtonClicked("*"))
        self.btndivide.clicked.connect(lambda: self.onButtonClicked("/"))
        self.btnequal.clicked.connect(self.onEqualClicked)
        self.btnclear.clicked.connect(self.onClearClicked)

        # Set the layout
        self.setLayout(vbox)

    # Define the button clicked slot
    def onButtonClicked(self, text):
        self.display.setText
        
    def onEqualClicked(self):
        try:
            result = eval(self.display.text())
            self.display.setText(str(result))
        except:
            self.display.setText("Error")

    # Define the clear button clicked slot
    def onClearClicked(self):
        self.display.clear()


app = QApplication(sys.argv)
calculator = Calculator()
calculator.show()
sys.exit(app.exec_())
```
